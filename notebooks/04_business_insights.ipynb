##***Business Insights***

####**🧠 1. Problem Statement**

The objective of this project is to build a predictive system to identify machine failures in advance using operational parameters such as temperature, torque, rotational speed, and tool wear.

Early prediction of failures enables:

Reduced downtime
Preventive maintenance
Cost optimization

**⚙️ 2. Model Selection Justification**

We experimented with multiple models:

Logistic Regression

Decision Tree

Random Forest

XGBoost

✅ Why XGBoost was selected as final model

XGBoost outperformed other models due to:

Better handling of non-linear relationships

Strong performance on imbalanced datasets

Built-in regularization (prevents overfitting)

Higher ROC-AUC (~0.89) compared to other models


| Model               | Strength                             | Limitation               |
| ------------------- | ------------------------------------ | ------------------------ |
| Logistic Regression | Simple, interpretable                | Poor on complex patterns |
| Decision Tree       | Easy to understand                   | Overfitting              |
| Random Forest       | Stable                               | Lower recall             |
| **XGBoost**         | **Best balance of recall & ROC-AUC** | Slightly complex         |




✅ Why 0.4 was chosen

Detects ~60% of failures

Acceptable trade-off with precision

Aligns with business priority: minimize missed failures

🧠 Business Interpretation

Better to inspect a healthy machine than miss a failing one

📊 Model Performance Summary

ROC-AUC: ~0.89

Recall (Failure): ~60%

Precision (Failure): ~0.29–0.44

Accuracy: ~97–98%

**🔍 Key Feature Insights (SHAP Analysis)**

Using SHAP, we identified important drivers of failure:

🔥 Top contributing features:

Tool Wear

Torque

Temperature Difference

Power

Torque-Speed Ratio

**📌 Key Observations**

High tool wear increases failure probability

High torque & power → mechanical stress

Temperature imbalance affects performance

**💼 Actionable Business Insights**

**🔧 1. Preventive Maintenance Scheduling**

Machines with high tool wear should be serviced early

**🌡️ 2. Temperature Monitoring**

Maintain optimal difference between:

Air temperature

Process temperature

**⚙️ 3. Load Management**

High torque & speed combinations increase failure risk

Avoid operating machines under extreme stress

**📊 4. Real-time Monitoring System**

Deploy model for:

Continuous monitoring

Early alerts

**💰 5. Business Impact Quantification**

Let’s assume:

1000 machines

10% failure rate → 100 failures

Cost per failure = ₹50,000

**🚨 Without model:**

Total loss = 100 × 50,000 = ₹50,00,000

**✅ With model (Recall = 60%)**

Detects 60 failures early

Prevents major damage

Savings = 60 × 50,000 = ₹30,00,000

**📈 Impact**

👉 ₹30 lakh potential savings
